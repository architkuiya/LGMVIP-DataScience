```python
import pandas as pd, matplotlib.pyplot as plt, seaborn as sns, sklearn as sk
%matplotlib inline
```

```python
df = pd.read_csv("Iris.csv")
```

```python
df.drop('Id',axis = 1,inplace=True)
df.head()
```

```python
sns.scatterplot(x=df['SepalLengthCm'],y=df['SepalWidthCm'],hue=df['Species'],palette=['green','red','blue'])
plt.legend(bbox_to_anchor=(1.35, 1))
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop('Species',axis=1)
y=df['Species']
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.3)
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier(criterion='gini',min_samples_split=2)
model.fit(xtrain, ytrain)
```

```python
model.score(xtest,ytest)
```

```python
fig = plt.figure(figsize=(25,20))
tree.plot_tree(model)
```

```python
predictions = model.predict(xtest)
predictions
from sklearn.metrics import accuracy_score
accuracy_score(ytest, predictions)
```
